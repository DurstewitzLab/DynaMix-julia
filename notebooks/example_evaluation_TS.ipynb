# Zero-shot TS forecasting using DynaMix


```julia
using DSR
using NPZ
```

## Load model

```julia
DynaMix, Observation_model = load_model("../models/dynamix-3d-alrnn-v1.0.bson"); # Add model path here
```

## Time Series Forecasting

```julia
# Forecasting Hyperparameters
# First forecast might take longer due to precompilation
CL = 200; # Context length
T = 200; # Prediction length

data = npzread("../notebooks/test_data/sine.npy");

context = data[1:CL,:];
ground_truth = data[CL+1:CL+T,:];

reconstruction = DynaMix_forecasting_pipeline(
                    DynaMix, Observation_model, context, T,
                    preprocessing_method="pos_embedding"
);

plot_TS_forecast(ground_truth, context, reconstruction[:,1],T)
```
